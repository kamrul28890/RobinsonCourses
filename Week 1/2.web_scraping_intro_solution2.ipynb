```python
#!pip install requests
#!pip install beautifulsoup4
#!pip install pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
titles = []
prices = []
descriptions = []
ratings = []
url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'
request = requests.get(url)
```

```python
#request.text
```

```python
soup = BeautifulSoup(request.text, "html.parser")

for product in soup.find_all('div', {'class': 'col-md-4 col-xl-4 col-lg-4'}):
    for pr in product.find_all('div', {'class': 'caption'}):
        for p in pr.find_all('h4', {'class': 'float-end price card-title pull-right'}):
            prices.append(p.text)
        for title in pr.find_all('a' , {'title'}):
            titles.append(title.get('title'))
        for desc in pr.find_all('p',{'class':'description'}):
            descriptions.append(desc.text)
    for rt in product.find_all('div', {'class': 'ratings'}):
        for rating in rt.find_all('p', attrs={'data-rating': True}):
            ratings.append(rating.get('data-rating'))
    
#add your own code to get the product description and product ratings, then export all information to csv
    
#build dataframe and export to csv            
product_df = pd.DataFrame(zip(titles,prices,descriptions,ratings), columns =['Titles', 'Prices', 'Description', 'Rating'])  
product_df.head()
```

```python
product_df.to_csv("ecommerce.csv",index=False)
```
